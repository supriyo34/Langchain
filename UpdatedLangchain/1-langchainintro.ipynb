## Option 2: Modern Agents (Recommended for LangChain v1.x)

In newer versions of LangChain, agents are constructed as state graphs under the hood. You can use the unified `create_agent` factory function from `langchain.agents`.

```python
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI

# Initialize Google's Gemini model
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Define a tool for the agent
def get_weather(location: str) -> str:
    """Return the weather forecast for the specified location."""
    return f"It's sunny and 72°F in {location}"

# Create the agent
agent = create_agent(
    model=model,
    tools=[get_weather],
    system_prompt="You are a helpful assistant."
)

# Run the agent and stream updates
inputs = {"messages": [{"role": "user", "content": "What is the weather in New York?"}]}
for chunk in agent.stream(inputs, stream_mode="updates"):
    print(chunk)
```
